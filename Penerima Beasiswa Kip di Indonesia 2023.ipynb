**Proses Cleaning**

```python
# Load data dengan mengabaikan baris pertama
df = pd.read_excel("beasiswa_kip.xlsx", skiprows=2)
```

```python
# Ubah nama kolom menjadi lebih deskriptif
df.columns = ["No", "Kode Kemdagri", "Kode BPS", "Provinsi", "Jumlah"]
```

```python
# Hapus kolom 'No' karena tidak  berguna
df.drop(columns=["No"], inplace=True)
```

```python
# Konversi ke tipe numerik
df["Jumlah"] = pd.to_numeric(df["Jumlah"], errors="coerce")
```

```python
# Hapus baris yang memiliki nilai NaN pada kolom 'Provinsi' atau 'Jumlah'
df = df.dropna(subset=["Provinsi", "Jumlah"])
```

```python
# Menghapus baris yang memiliki nilai NaN di kolom mana pun
df_cleaned = df.dropna()

# Menampilkan hasil setelah pembersihan
print(df_cleaned)
```

# EDA

##  Statistik Dasar Penerima Beasiswa KIP Kuliah

```python
print(df_cleaned["Jumlah"].describe())
```

Jumlah mahasiswa penerima Beasiswa KIP Kuliah tahun 2023 di Indonesia menunjukkan distribusi yang cukup bervariasi, dengan nilai minimum 2.272 dan maksimum 75.951 penerima di suatu wilayah, serta rata-rata sebesar 17.258 penerima. Tingginya standar deviasi (18.066) mengindikasikan adanya ketimpangan dalam distribusi beasiswa, yang kemungkinan dipengaruhi oleh faktor selain dunia pendidikan, seperti tingkat kemiskinan regional, aksesibilitas internet, dan kondisi infrastruktur di masing-masing daerah. Wilayah dengan jumlah penerima tinggi cenderung memiliki populasi mahasiswa lebih besar atau tingkat keterjangkauan pendidikan yang lebih rendah, sehingga kebutuhan akan bantuan finansial lebih mendesak. Selain itu, kebijakan pemerataan dari pemerintah, termasuk pembangunan ekonomi dan ketersediaan lapangan kerja di daerah tertentu, juga dapat berkontribusi terhadap variasi penerimaan beasiswa, karena daerah dengan tingkat pengangguran tinggi kemungkinan memiliki lebih banyak mahasiswa yang mengandalkan bantuan finansial untuk melanjutkan pendidikan tinggi.

## Provinsi dengan Penerima Beasiswa Terbanyak & Tersedikit

```python
top_7 = df_cleaned.nlargest(7, 'Jumlah')
print(top_7)
```

```python
bottom_7 = df_cleaned.nsmallest(7, 'Jumlah')
print(bottom_7)
```

```python
# Buat pie chart
import matplotlib.pyplot as plt

top_7 = df_cleaned.nlargest(7, 'Jumlah')
plt.figure(figsize=(8, 8))
plt.pie(top_7['Jumlah'], labels=top_7['Provinsi'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribusi 7 Provinsi dengan Jumlah Penerima Terbanyak')
plt.show()
```

Data menunjukkan distribusi penerima manfaat yang tidak hanya dipengaruhi oleh aspek pendidikan, tetapi juga oleh faktor sosial ekonomi yang luas. Misalnya, provinsi seperti Jawa Timur dan Jawa Barat memiliki persentase penerima manfaat tertinggi, sekitar 23,5% dan 22,7%, yang mungkin mencerminkan populasi yang lebih besar, tingkat kemiskinan, atau akses terhadap layanan kesehatan dan lapangan pekerjaan yang lebih baik. Ketimpangan dalam pembangunan antarprovinsi, termasuk kesiapan infrastruktur dan investasinya dalam fasilitas publik, juga berkontribusi pada variasi ini. Oleh karena itu, perencanaan kebijakan harus mempertimbangkan integrasi berbagai sektor untuk mendukung pertumbuhan yang lebih merata di seluruh provinsi.

```python
# Buat pie chart
import matplotlib.pyplot as plt

top_7 = df_cleaned.nsmallest(7, 'Jumlah')
plt.figure(figsize=(8, 8))
plt.pie(top_7['Jumlah'], labels=top_7['Provinsi'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribusi 7 Provinsi dengan Jumlah Penerima Terdikit')
plt.show()
```

Visualisasi ini menunjukkan ketimpangan penerima Beasiswa KIP Kuliah, di mana 7 provinsi dengan jumlah penerima terendah, seperti Gorontalo (19,5%) dan Papua Barat (19,1%), masih menghadapi keterbatasan akses pendidikan. Faktor seperti infrastruktur, jumlah perguruan tinggi, dan kondisi ekonomi berperan dalam rendahnya penerimaan beasiswa di wilayah ini. Jika tidak segera diatasi, kesenjangan ini dapat memperlambat pembangunan sumber daya manusia dan memperlebar disparitas ekonomi. Oleh karena itu, diperlukan kebijakan yang lebih inklusif, seperti peningkatan alokasi beasiswa dan penguatan infrastruktur pendidikan di daerah tertinggal.

## Distribusi Penerima Beasiswa KIP Kuliah dalam Visualisasi

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Urutkan data berdasarkan jumlah penerima
df_sorted = df_cleaned.sort_values(by='Jumlah', ascending=False)

# Buat figure
plt.figure(figsize=(12, 6))
sns.barplot(x=df_sorted['Provinsi'], y=df_sorted['Jumlah'], palette="Blues_r")

# Rotasi label x agar lebih terbaca
plt.xticks(rotation=90)

# Tambahkan judul dan label
plt.title('Distribusi Penerima Beasiswa KIP Kuliah per Provinsi')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah Penerima')

# Tampilkan plot
plt.show()
```

Distribusi penerima beasiswa KIP Kuliah menunjukkan konsentrasi yang signifikan di pulau Jawa, khususnya Jawa Timur, yang mencerminkan tidak hanya faktor aksesibilitas pendidikan tetapi juga aspek sosial ekonomi, seperti urbanisasi dan perkembangan infrastruktur. Provinsi dengan jumlah penerima yang lebih rendah, seperti Kepulauan Bangka Belitung dan Kalimantan Utara, mungkin menghadapi tantangan dalam hal akses pendidikan, ketidakmerataan ekonomi, dan rendahnya investasi dalam pengembangan sumber daya manusia. Selain itu, variabel lain seperti dukungan pemerintah daerah, budaya lokal terhadap pendidikan, dan keterlibatan sektor swasta dalam pendanaan pendidikan tinggi juga dapat mempengaruhi angka penerima beasiswa, menciptakan disparitas yang perlu ditangani secara holistik untuk mencapai pemerataan dalam akses pendidikan di seluruh Indonesia.

## Perbandingan Penerima Beasiswa di Wilayah Timur vs Barat (Bar Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Provinsi": ["Aceh", "Sumatera Utara", "Sumatera Barat", "Riau", "Jambi", "Sumatera Selatan", "Bengkulu", "Lampung", "Kepulauan Bangka Belitung", "Kepulauan Riau", "D.K.I. Jakarta", "Jawa Barat", "Jawa Tengah", "D.I. Yogyakarta", "Jawa Timur", "Banten", "Bali", "Nusa Tenggara Barat", "Nusa Tenggara Timur", "Kalimantan Barat", "Kalimantan Tengah", "Kalimantan Selatan", "Kalimantan Timur", "Kalimantan Utara", "Sulawesi Utara", "Sulawesi Tengah", "Sulawesi Selatan", "Sulawesi Tenggara", "Gorontalo", "Sulawesi Barat", "Maluku", "Maluku Utara", "Papua", "Papua Barat"],
    "Jumlah": [29607, 40709, 23639, 12071, 7044, 15733, 7573, 14281, 2272, 3294, 12540, 73396, 52934, 11078, 75951, 14356, 6941, 18935, 20322, 10740, 4398, 8976, 10562, 2878, 10893, 10218, 27071, 12076, 5839, 9799, 8581, 5483, 10886, 5712]
}

df = pd.DataFrame(data)

# Pembagian Wilayah
wilayah_barat = ["Aceh", "Sumatera Utara", "Sumatera Barat", "Riau", "Jambi", "Sumatera Selatan", "Bengkulu", "Lampung", "Kepulauan Bangka Belitung", "Kepulauan Riau", "D.K.I. Jakarta", "Jawa Barat", "Jawa Tengah", "D.I. Yogyakarta", "Jawa Timur", "Banten"]
wilayah_timur = ["Bali", "Nusa Tenggara Barat", "Nusa Tenggara Timur", "Kalimantan Barat", "Kalimantan Tengah", "Kalimantan Selatan", "Kalimantan Timur", "Kalimantan Utara", "Sulawesi Utara", "Sulawesi Tengah", "Sulawesi Selatan", "Sulawesi Tenggara", "Gorontalo", "Sulawesi Barat", "Maluku", "Maluku Utara", "Papua", "Papua Barat"]

# Menghitung total penerima beasiswa
total_barat = df[df["Provinsi"].isin(wilayah_barat)]["Jumlah"].sum()
total_timur = df[df["Provinsi"].isin(wilayah_timur)]["Jumlah"].sum()

# Visualisasi
labels = ["Wilayah Barat", "Wilayah Timur"]
values = [total_barat, total_timur]

plt.figure(figsize=(8, 5))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['blue', 'orange'], startangle=90, shadow=True)
plt.title("Perbandingan Penerima Beasiswa Wilayah Timur vs Barat")
plt.show()
```

grafik ini menunjukkan ketimpangan yang signifikan dalam penerimaan beasiswa antara wilayah Barat (67,6%) dan Timur (32,4%). Faktor-faktor di luar dunia pendidikan, seperti perekonomian regional, infrastruktur, dan aksesibilitas layanan sosial, berkontribusi terhadap perbedaan ini. Wilayah Barat kemungkinan lebih berkembang dari segi ekonomi, dengan lebih banyak peluang kerja yang mendukung keberlanjutan pendidikan, serta infrastruktur yang lebih baik yang memudahkan akses ke lembaga pendidikan. Sementara itu, wilayah Timur mungkin menghadapi berbagai tantangan, mulai dari keterbatasan sumber daya hingga kurangnya perhatian pembangunan, yang berdampak langsung pada kesempatan masyarakat untuk mengakses bantuan pendidikan. Pembenahan dalam kebijakan dan investasi di wilayah Timur sangat diperlukan untuk mengurangi kesenjangan tersebut.
